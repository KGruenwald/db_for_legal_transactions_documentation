```python


import os
from bs4 import BeautifulSoup
import pandas as pd
import re

ns = {'tei': 'http://www.tei-c.org/ns/1.0', "xml": "http://www.w3.org/XML/1998/namespace"}


def readlist(pathtofile, filename):
    with open(pathtofile+"/"+filename, encoding="utf-8") as infile:
        blob = infile.read()
        soup = BeautifulSoup(blob, 'lxml')
        return soup.findAll('person')


def readfiles(pathtofiles, personlistnodes):
    #print(personlist)
    personsFrame = pd.DataFrame()
    personsDict = {}
    personslist = []
    personsref = []
    personstext = []
    personvorkommen = []
    personfiles = []
    personslistvorkommen = []
    #print(personlistnodes)
    test = personlistnodes.findAll("xml:id"="QGW_II_I_register_pg_545")
    print(test)
    for filename in os.listdir(pathtofiles):
       with open(pathtofiles+"/"+filename, encoding="utf-8") as infile:
           blob = infile.read()
           soup = BeautifulSoup(blob, 'lxml')
           actualregest = soup.find("idno").text
           datenode = soup.find("creation").find("date")
           date = createdate(datenode)
           personNodes = soup.findAll(type="person")
           for rs in personNodes:
               text = ' '.join(rs.text.replace("\n", "").split())
               getpersonroles(rs)
               getpersonevent(rs)
               if rs.has_attr('ref'):
                   ref = rs['ref']
                   createperson(ref, personlistnodes)
                   
                #    ref = rs['ref']
                 #   if ref not in personsDict:
                  #      sublist = [text, 1, [filename]]
                   #     personsDict[ref] = sublist
                    #else:
                     #   personsDict[ref][1] += 1
                      #  if filename not in personsDict[ref][2]:
                       #     personsDict[ref][2].append(filename)
               #else:
                #   personslist.append(text)
                 #  personslistvorkommen.append(filename)


    for key, item in personsDict.items():
        personsref.append(key)
        personstext.append(item[0])
        personvorkommen.append(item[1])
        filestring = ", ".join(item[2])
        personfiles.append(filestring)

    for entry in personslist:
        personsref.append('')
        personstext.append(entry)
        personvorkommen.append(1)

    for entry in personslistvorkommen:
        personfiles.append(entry)


    personsFrame['ref'] = personsref
    personsFrame['text'] = personstext
    personsFrame['anzahl'] = personvorkommen
    personsFrame['xml'] = personfiles

    personsFrame.to_csv("../data/output/extrectedPersons.csv", sep=";", encoding="utf-8")



def getpersonroles(persnode):
    roles = {}
    rolenamenodes = persnode.findAll("rolename")
    if rolenamenodes is not None:
        roles_prof = []
        roles_title = []
        roles_ref = []
        for rolenamenode in rolenamenodes:
            content = re.sub('\s\s+',' ',rolenamenode.text)
            if rolenamenode.has_attr('type'):
                roletype = rolenamenode['type']
                if roletype == "prof":
                    roles_prof.append(content)
                elif roletype == "title":
                    roles_title.append(content)
                elif roletype == "ref":
                    roles_ref.append(content)
    
    roles['prof'] = roles_prof
    roles['title'] = roles_title
    roles['ref'] = roles_ref
    
    return roles
  
def getpersonevent(persnode):
    eventnode = getparentevent(persnode)
    if eventnode is not None:
        event_ref = eventnode['ref']
        return event_ref
    else: return None


def getparentevent(node):
    actualnode = node
    while actualnode.parent.name is not "p":
        actualnode = actualnode.parent
        if actualnode.name is "rs":
            if actualnode['type'] == "event":
                return actualnode
    return None

def createdate(node):
    when = node["when"]
    return {"year": when[:4], "month": when[4:6], "day" : when[6:8] }


def createperson(ref, personlistnodes):
    ref = ref.replace("#", '')
    print(ref)
    #print(personlistnodes)
    #print(personnode)
    

personlistnodes = readlist("../../data/lists", "personList.xml")    
readfiles("../../data/done", personlistnodes)
```
