```python




import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

def mergeTable(csv, mergeover):

    csv = pd.read_csv(csv, encoding="ISO-8859-1", delimiter="\t")


    filteredlist = csv[mergeover]
    filteredset = set(filteredlist)
    filteredlist = list(filteredset)

    newlist = []

    for item in filteredlist:
        filter = csv[mergeover] == item
        filteredlist = csv[filter]
        id = filteredlist['ID'].tolist()[0]
        type = filteredlist['type'].tolist()[0]
        name_reg = str(filteredlist['reg'].tolist()[0])
        name_orig = str(filteredlist['orig'].tolist()[0])
        print(name_reg)
        print(name_orig)
        name = name_reg + " | " + name_orig
        anzahl = len(filteredlist[mergeover])

        linkinglist = filteredlist['linking'].tolist()
        countref = linkinglist.count('ref')
        countcorresp = linkinglist.count('corresp')


        years = filteredlist['date'].astype(str).replace({np.nan: "None"})
        years = ", ".join(years.tolist()).replace(".0", "")
        xmls = ", ".join(filteredlist['xml'].tolist())

        sublist =[id,  name, type, anzahl, countref, countcorresp, xmls, years]
        newlist.append(sublist)
    writetotable(newlist)


def writetotable(newlist):
    newFrame = pd.DataFrame()
    id =[]
    name =[]
    type = []
    anzahl = []
    ref_count =[]
    corresp_count=[]
    xml =[]
    year = []

    for entry in newlist:
        id.append(entry[0])
        name.append(entry[1])
        type.append(entry[2])
        anzahl.append(entry[3])
        ref_count.append(entry[4])
        corresp_count.append(entry[5])
        xml.append(entry[6])
        year.append(entry[7])

    newFrame['ID'] = id
    newFrame['name_reg | name_orig'] = name
    newFrame['type'] = type
    newFrame['anzahl'] = anzahl
    newFrame['ref_count'] = ref_count
    newFrame['corresp_count'] = corresp_count
    newFrame['xml'] = xml
    newFrame['year'] = year

    newFrame.to_csv("../../data/output/csv/MergedAllPlaces.csv", sep="\t", encoding="utf-8")
    writer = pd.ExcelWriter('../data/output/xlsx/MergedAllPlaces.xlsx')
    newFrame.to_excel(writer, 'AllPlaces', index=False)
    writer.save()




mergeTable("../../data/output/csv/AllPlaces.csv", 'ID')

```
